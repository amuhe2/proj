## Machine Learning (Supervised/Unsupervised)

```python
import pandas as pd
from sklearn import svm
from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
```

```python
def prepare_data(df, train=True):
    if train:
        X, y = df.drop('Cover_Type', 1), df['Cover_Type']
    else:
        X, y = df, None
    return X, y


X_train, y_train = prepare_data(pd.read_csv('C:\Users\muheeb\Documents\IPython Notebooks\\train.csv', index_col=0))
X_test, y_test = prepare_data(pd.read_csv('C:\Users\muheeb\Documents\IPython Notebooks\\test.csv', index_col=0), train=False)

model = RandomForestClassifier(n_estimators=600,
                             max_features=13,
                             min_samples_split=1,
                             n_jobs=1,
                             random_state=10)
debug = False

if debug:
    scores = cross_val_score(model, X_train, y_train, cv=3)
    print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
else:
    model = model.fit(X_train, y_train)
    output = model.predict(X_test).astype(int)
    result = pd.Series(output, index=X_test.index)
    result.to_csv('output.csv', header=['Cover_Type'])
    print "please check the output.csv file in the folder"
```

```python
    scores = cross_val_score(model, X_train, y_train, cv=3)
    print('Accuracy: ' , scores.mean())
```

###### We have achieved an Accuracy of .7923 with random forest
